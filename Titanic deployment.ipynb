```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
#Train dataset
```

```python
datatrain = pd.read_csv('Nisarg Rade - train.csv')
datatrain.head(10)
```

```python
datatrain.info()
```

```python
datatrain = datatrain.drop(columns=['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'])#dropping coloumns
datatrain.Age.describe()
```

```python
datatrain.Age.fillna(datatrain.Age.mean(), inplace=True)#filling null values
datatrain.isnull().sum()
```

```python
datatrain.head(10)
```

```python
l_sex_dummies=pd.get_dummies(datatrain['Sex'],drop_first=True)
datatrain= pd.concat([datatrain,l_sex_dummies],axis=1)
datatrain.drop(['Sex'], axis=1, inplace=True )
datatrain.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
ssc=StandardScaler()
```

```python
feature_scale = ['Age','Fare']
datatrain[feature_scale] = ssc.fit_transform(datatrain[feature_scale])
datatrain.head(20)
```

```python
X=datatrain.drop(['Survived'],axis=1)
y=datatrain['Survived']
```

Support Vector Machines

```python
from sklearn.svm import SVC
model_svc = SVC( C= 100,kernel='rbf')
model_svc.fit(X, y)
```

Test Data

```python
datatest = pd.read_csv('Nisarg Rade - test.csv')
datatest.head(15)
```

```python
datamodel = datatest.drop(['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'], axis=1 )#droping colunns
datamodel.isnull().sum()
```

```python
datamodel['Age'].fillna(datamodel['Age'].mean(),inplace=True)
datamodel['Fare'].fillna(datamodel['Fare'].mean(),inplace=True)
```

```python
datamodel.info()
```

```python
l_sex_dummies=pd.get_dummies(datamodel['Sex'],drop_first=True)
datamodel= pd.concat([datamodel,l_sex_dummies],axis=1)
datamodel.drop(['Sex'], axis=1, inplace=True )
datamodel[feature_scale] = ssc.fit_transform(datamodel[feature_scale])
datamodel.head()
```

```python
y_predicted = model_svc.predict(datamodel)
```

```python
result = pd.DataFrame({
        "PassengerId": datatest['PassengerId'],
        "Survived": y_predicted
    })
```

```python
result.to_csv('titanic_result_deployed.csv', index=False)
```
